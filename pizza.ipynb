```python
import matplotlib.pyplot as plt

#os dois parametros como 'sabores' e quantidade de 'amor'
sabores = ['Calabresa', 'Strogonoff', 'Portuguesa', '4 queijos']
amor = [40, 50, 1, 9]

#as cores
cores = ['#ff9999', '#ffcc99', '#99ff99', '#66b3ff']

#montando a pizza o 'amor' vai ser a quantidade, o 'label=sabores' o nome de cada parte, e 'autopct='%1.1f%%' é a formatação da porcentagem
plt.pie(amor, labels=sabores, colors=cores, autopct='%1.1f%%')

#o nome do grafico
plt.title('Pizas')

plt.show()
```
